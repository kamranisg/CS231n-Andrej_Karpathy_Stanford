```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
```

```python
(x_train,y_train),(x_test,y_test)=cifar10.load_data()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=plt.figure(figsize=(6,6))
for i in range (1,26):
    fig.add_subplot(5,5,i)
    plt.imshow(x_train[i])
```

```python
a=np.array(x_train.reshape(50000,3072))
b=a-np.array(x_test[1].reshape(1,3072))
c=np.sum(b,axis=1)
index=np.argpartition(c,10)
s=c[index[:10]]
fig=plt.figure(figsize=(60,60))

fig.add_subplot(2,12,1)
plt.imshow(x_test[1])
for i in range(1,11):
    fig.add_subplot(2,12,i+1)
    d=np.where(c==s[i-1])
    plt.imshow(x_train[d[0]][0])
```

```python
a2=np.array(x_train.reshape(50000,3072))
b2=a2-np.array(x_test[3].reshape(1,3072))
c2=np.sum(b2,axis=1)
index2=np.argpartition(c2,10)
s2=c2[index2[:10]]
fig=plt.figure(figsize=(60,60))

fig.add_subplot(2,12,1)
plt.imshow(x_test[3])
for i in range(1,11):
    fig.add_subplot(2,12,i+1)
    d2=np.where(c2==s2[i-1])
    plt.imshow(x_train[d2[0]][0])
```
